```
def input_matrix_2x2(nama):
    print(f"Input elemen {nama} (2x2):")
    return [[int(input(f"{nama}[{i}][{j}]: ")) for j in range(2)] for i in range(2)]

def tambah(A, B):
    return [[A[i][j] + B[i][j] for j in range(2)] for i in range(2)]

def kali(A, B):
    return [[A[i][0]*B[0][j] + A[i][1]*B[1][j] for j in range(2)] for i in range(2)]

def transpose(A):
    return [[A[j][i] for j in range(2)] for i in range(2)]

while True:
    print("\n=== MENU KASUS 1 ===")
    print("1. Penjumlahan")
    print("2. Perkalian")
    print("3. Transpose")
    print("4. Keluar")
    pilih = input("Pilihan: ")

    if pilih == "1":
        A = input_matrix_2x2("A")
        B = input_matrix_2x2("B")
        print("Hasil penjumlahan:")
        for r in tambah(A, B): print(r)

    elif pilih == "2":
        A = input_matrix_2x2("A")
        B = input_matrix_2x2("B")
        print("Hasil perkalian:")
        for r in kali(A, B): print(r)

    elif pilih == "3":
        A = input_matrix_2x2("A")
        print("Transpose matriks A:")
        for r in transpose(A): print(r)

    elif pilih == "4":
        break
    else:
        print("Pilihan tidak valid!")
```

```
45
def input_matrix(m, n, nama):
    print(f"Input elemen {nama} ({m}x{n}):")
    return [[int(input(f"{nama}[{i}][{j}]: ")) for j in range(n)] for i in range(m)]

def tambah_matriks(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def kali_matriks(A, B):
    hasil = []
    for i in range(len(A)):
        baris = []
        for j in range(len(B[0])):
            baris.append(sum(A[i][k]*B[k][j] for k in range(len(B))))
        hasil.append(baris)
    return hasil

def transpose(A):
    return [[A[j][i] for j in range(len(A))] for i in range(len(A[0]))]

while True:
    print("\n=== MENU KASUS 2 ===")
    print("1. Penjumlahan dua matriks")
    print("2. Perkalian dua matriks")
    print("3. Transpose matriks")
    print("4. Keluar")
    pilih = input("Pilihan: ")

    if pilih == "1":
        print("Ukuran A: 2x3, Ukuran B: 3x2 → Tidak dapat dijumlahkan!")

    elif pilih == "2":
        A = input_matrix(2, 3, "A")
        B = input_matrix(3, 2, "B")
        print("Hasil perkalian A × B :")
        for r in kali_matriks(A, B): print(r)

    elif pilih == "3":
        A = input_matrix(2, 3, "A")
        print("Transpose A:")
        for r in transpose(A): print(r)

    elif pilih == "4":
        break
    else:
        print("Pilihan tidak valid!")
```
